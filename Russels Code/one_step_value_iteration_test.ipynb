```python
import numpy as np
import matplotlib.pyplot as plt

def plot_two_lines():
    # Generate x values
    x = np.linspace(0, 10, 100)

    # Define the parameters for the first line
    slope1 = 2
    intercept1 = 3

    # Define the parameters for the second line
    slope2 = -1
    intercept2 = 6

    # Calculate the y values for the first line
    y1 = slope1 * x + intercept1

    # Calculate the y values for the second line
    y2 = slope2 * x + intercept2

    # Plot the lines
    plt.plot(x, y1, label='Line 1')
    plt.plot(x, y2, label='Line 2')

    # Add labels and a legend
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Display the plot
    plt.show()

# Call the function to create the plot
plot_two_lines()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_two_lines():
    # Define alpha
    alpha = 0.9

    # Generate beta values
    beta = np.arange(0.1, 1.1, 0.1)

    # Calculate x values
    x = beta * alpha

    # Define the parameters for the first line
    slope1 = 2
    intercept1 = 3

    # Define the parameters for the second line
    slope2 = -1
    intercept2 = 6

    # Calculate the y values for the first line
    y1 = slope1 * x + intercept1

    # Calculate the y values for the second line
    y2 = slope2 * x + intercept2

    # Plot the lines
    plt.plot(x, y1, label='Line 1')
    plt.plot(x, y2, label='Line 2')

    # Add labels and a legend
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Display the plot
    plt.show()

# Call the function to create the plot
plot_two_lines()
```
