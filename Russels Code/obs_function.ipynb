# Observation Functions

This is just to play around with the idea of having random observations and drawing a sample from those. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate observation values
def obs_val(mu, sigma, n):
    obsVal = np.random.normal(mu, sigma, n)
    return obsVal

# Function to plot the experiment results
def plot_experiment(obsVal):
    plt.figure()
    plt.hist(obsVal, bins=20, color='blue', alpha=0.7)
    plt.xlabel('Observation Values')
    plt.ylabel('Frequency')
    plt.title('Experiment Results')
    plt.grid(True)
    plt.show()

# Parameters for the experiment
mu = 0.5
sigma = 0.2
num_samples = 1000

# Generate observation values
obsValues = obs_val(mu, sigma, num_samples)

# Plot the experiment results
plot_experiment(obsValues)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to generate observation values
def obs_val(mu, sigma, n):
    obsVal = np.random.normal(mu, sigma, n)
    return obsVal

# Function to plot the experiment results
def plot_experiment(obsVal, mu, sigma):
    plt.figure()
    
    # Plot histogram
    plt.hist(obsVal, bins=10, density=True, color='blue', alpha=0.5, label='Observation Values')
    
    # Plot normal distribution
    x = np.linspace(min(obsVal), max(obsVal), 100)
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, y, color='red', label='Normal Distribution')
    
    plt.xlabel('Observation Value')
    plt.ylabel('Frequency / Probability Density')
    plt.title('Experiment Results')
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters for the experiment
mu = 0.5
sigma = 0.2
num_samples = 100

# Generate observation values
obsValues = obs_val(mu, sigma, num_samples)

# Plot the experiment results
plot_experiment(obsValues, mu, sigma)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to generate observation probabilities
def obs_prob(mu, sigma, obs_val):
    obsProb = norm.pdf(obs_val, loc=mu, scale=sigma)
    return obsProb

# Function to plot the experiment results
def plot_experiment(obsProb, obsVal, mu, sigma):
    plt.figure()

    # Plot observation probabilities
    plt.plot(obsVal, obsProb, 'bo', label='Observation Probabilities')

    # Plot normal distribution
    x = np.linspace(min(obsVal), max(obsVal), 100)
    y = norm.pdf(x, loc=mu, scale=sigma)
    plt.plot(x, y, 'r-', label='Normal Distribution')

    plt.xlabel('Observation Value')
    plt.ylabel('Probability')
    plt.title('Experiment Results')
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters for the experiment
mu = 0.5
sigma = 0.2
num_samples = 100

# Generate observation values
obsValues = obs_val(mu, sigma, num_samples)

# Generate observation probabilities
obsProbs = obs_prob(mu, sigma, obsValues)

# Plot the experiment results
plot_experiment(obsProbs, obsValues, mu, sigma)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def obs_val(mu, sigma, num_samples):
    obsVal = np.random.normal(mu, sigma, num_samples)
    return obsVal

def obs_prob(obsVal, mu, sigma_att):
    obsProb = norm.pdf(obsVal, loc=mu, scale=sigma_att)
    return obsProb

# Parameters
mu = 1
sigma = 1
num_samples = 1000

# Generate observation values
obsVal = obs_val(mu, sigma, num_samples)

# Calculate observation probabilities
sigma_att = 0.5
obsProb = obs_prob(obsVal, mu, sigma_att)

# Plot histogram of observation values
plt.figure(figsize=(8, 6))
plt.hist(obsVal, bins=30, density=True, alpha=0.7, color='steelblue')
plt.xlabel('Observation Values')
plt.ylabel('Frequency')
plt.title('Histogram of Observation Values')

# Plot observation probabilities
plt.plot(obsVal, obsProb, 'ro', linewidth=2, label='Observation Probabilities')

# Plot normal distribution curve
x = np.linspace(mu - 3*sigma_att, mu + 3*sigma_att, 100)
y = norm.pdf(x, loc=mu, scale=sigma_att)
plt.plot(x, y, 'ko', linewidth=2, label='Normal Distribution')

plt.legend()
plt.grid(True)
plt.show()
```
